```python
import time

import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
```

```python
df_train = pd.read_csv("train_set.csv")
df_train
```

```python
df_test = pd.read_csv("test_set.csv")
df_test
```

```python
df_train = df_train[[
    "duration", "trip_route_category", "start_station", "end_station", "passholder_type"
]]
df_train
```

```python
df_test = df_test[[
    "duration", "trip_route_category", "start_station", "end_station"
]]
df_test
```

```python
# eliminar NaN por filas
df_train = df_train.dropna()
```

```python
# index de 0 a n-1
df_train = df_train.reset_index(drop=True)
```

```python
ohe = OneHotEncoder(sparse=False)
transformed_train = ohe.fit_transform(df_train[["trip_route_category"]])
ohe.categories_
```

```python
ohe = OneHotEncoder(sparse=False)
transformed_test = ohe.fit_transform(df_test[["trip_route_category"]])
ohe.categories_
```

```python
df_train["one_way"] = transformed_train[:,0]
df_train["round_trip"] = transformed_train[:,1]
df_train = df_train.drop(["trip_route_category"], axis=1)

df_test["one_way"] = transformed_test[:,0]
df_test["round_trip"] = transformed_test[:,1]
df_test = df_test.drop(["trip_route_category"], axis=1)
```

```python
df_train["one_way"] = df_train["one_way"].apply(lambda x: int(x))
df_train["round_trip"] = df_train["round_trip"].apply(lambda x: int(x))

df_test["one_way"] = df_test["one_way"].apply(lambda x: int(x))
df_test["round_trip"] = df_test["round_trip"].apply(lambda x: int(x))
```

```python
le = LabelEncoder().fit(df_train["passholder_type"])
df_train["passholder_type_target"] = le.transform(df_train["passholder_type"])
df_train = df_train.drop(["passholder_type"], axis=1)
```

```python
x_train = df_train.iloc[:, :-1].to_numpy()
```

```python
x_test = df_test.to_numpy()
```

```python
y_train = df_train["passholder_type_target"].to_numpy()
```

```python
inicio = time.time()

clf = RandomForestClassifier()
clf.fit(x_train, y_train)

print("Elapsed:", time.time() - inicio)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_pred_name = le.inverse_transform(y_pred)
```

```python
df_pred = pd.read_csv("sample_submission.csv")
df_pred
```

```python
df_pred["passholder_type"] = y_pred_name
df_pred
```

```python
df_pred.to_csv("submission.csv", index=False)
```
